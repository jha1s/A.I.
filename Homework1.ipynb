##__4.   자율 주행__

   * __Waymo(웨이모)__
     * 구글 슬렉스의 연구소에서 개발하는 무인 자동차(구글 카)
     * 미국 텍사스의 피닉스에서 waymo one이라는 로보 택시를 시범적으로 운영 중

   * __Tesla Autopilot(테슬라 오토파일럿)__
      * 테슬라 자동차에서 사용하는 ADAS<sup id="a8">[2](#footnote8)</sup> 시스템
      * 가장 많은 자율 주행 자동차를 판매(2019년말 기준 70만대)
      * 딥러닝에서 가장 중요시되는 데이터를 가장 많이 축적(약 48억km)
      * Lv 2의 부분 자동화가 가능하며 FSD옵션을 넣으면 Lv 4 고도 자동화까지 가능
